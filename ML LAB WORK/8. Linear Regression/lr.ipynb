```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from matplotlib import pyplot as plt
```

```python
train = pd.read_csv('train.csv')
test =  pd.read_csv('test.csv')

train_x = train['x']
test_x = test['x']
train_y = train['y']
test_y = test['y']
```

```python
plt.subplots(figsize=(15,6))
plt.suptitle("Distribution of Data")

plt.subplot(1,2,1)
plt.scatter(train_x,train_y,c='blue',edgecolors='white')
plt.xlabel("Input Training Samples")
plt.ylabel("Output Training Samples")

plt.subplot(1,2,2)
plt.scatter(test_x,test_y,c='red',edgecolors='white')
plt.xlabel("Input Testing Samples")
plt.ylabel("Output Testing Samples")

```

```python
train_x_transformed = np.array(train_x).reshape(-1,1)
test_x_transformed = np.array(test_x).reshape(-1,1)
```

```python
lr = LinearRegression()

model = lr.fit(train_x_transformed,train_y)
```

```python
y_pred = model.predict(test_x_transformed)
```

```python
plt.scatter(x=test_x,y=test_y,c='red',edgecolors='white')

plt.plot(test_x_transformed,y_pred,c='black')
plt.title("Line of best fit")
plt.xlabel("Input sample")
plt.ylabel("Predicted Values")
```

```python
print("The accuracy of the model is: ",model.score(test_x_transformed,test_y))
print("The mean squared error is: ",mean_squared_error(test_y,y_pred))
print("The mean absolute error is: ",mean_absolute_error(test_y,y_pred))
print("The slope of the line is: ",model.coef_[0])
print("The intercept of the line is is: ",model.intercept_)
```
